Importing pandas and the csv file

```python
import pandas as pd
df = pd.read_csv("vgsales.csv")
```

Limiting the data to the first 1000 rows

```python
df = df.head(1000)
```

Inspecting the dataset before cleaning

```python
df.info()
```

```python
df.describe()
```

Standardizing column names by striping spaces, lower-case, and replacing spaces with underscores. This is being done because clean column names are easier to type and access programmatically.


```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

Checking for any nulls

```python
df.isnull().sum()
```

Temporarily converting year to numeric to make it easier to drop rows where year is missing. Year is then converted to int. This is being done to enable time-based analysis.

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df = df.dropna(subset=['year']).copy()
df['year'] = df['year'].astype(int)
```

Cleaning the string columns using strip. This is done to remove whitespace inconsistencies and standardize formatting

```python
df['name'] = df['name'].str.strip()
df['platform'] = df['platform'].str.upper().str.strip()
df['genre'] = df['genre'].str.title().str.strip()
df['publisher'] = df['publisher'].str.strip()
```

Inspecting the dataset after performing cleaning operations

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

Saving the cleaned dataset as a new csv

```python
df.to_csv("vgsales_cleaned.csv", index=False)
```
